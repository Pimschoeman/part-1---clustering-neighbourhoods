Importing the pandas library for creating the dataframe

```python
import pandas as pd
```

reading the wikipedia table into a dataframe

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df.head(10)
```

creating a new dataframe which excludes the rows where borough == "not assigned"

```python
# i assumed that i could make a new dataframe with specifying that "not assigned" would not appear in column borough. 
dfwna= df[df.Borough != 'Not assigned']
```

```python
dfwna.head(10)
```

Modifying the dataframe so values would be combined

```python
dfwna =dfwna.groupby(['Postcode','Borough'])
dfwna=dfwna['Neighbourhood'].apply(', '.join).reset_index() 
dfwna.head(10)
```

Iterating through the dataframe and assigning the borough value to the neighbourhood value

```python
for i in range(len(dfwna)):
    if dfwna['Neighbourhood'].iloc[i] =='Not assigned':
        dfwna['Neighbourhood'].iloc[i]= dfwna['Borough'].iloc[i]
        
dfwna
```
